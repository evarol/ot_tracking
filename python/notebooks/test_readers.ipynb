```python
import numpy as np
import matplotlib.pyplot as plt

import readers
```

### Zimmer reader

```python
zimmer_path = '/home/mn2822/Desktop/WormOT/data/zimmer/mCherry_v00065-01581.hdf5'

with readers.ZimmerReader(zimmer_path) as reader:
    
    print(f't_start: {reader.t_start}')
    print(f't_stop: {reader.t_stop}')
    print(f'num frames: {reader.num_frames}')
    
    frame = reader.get_frame(500)
    plt.imshow(np.max(frame, 2).T);
```
