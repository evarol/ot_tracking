```python
import numpy as np
import matplotlib.pyplot as plt

from otimage import io 
```

### Zimmer reader

```python
zimmer_path = '/home/mn2822/Desktop/WormOT/data/zimmer/raw/mCherry_v00065-00115.hdf5'

with io.ZimmerReader(zimmer_path) as reader:
    
    print(f't_start: {reader.t_start}')
    print(f't_stop: {reader.t_stop}')
    print(f'num frames: {reader.num_frames}')
    print(f'units: {reader.units}')
    
    frame = reader.get_frame(25)
    plt.imshow(np.max(frame, 2).T);
```
